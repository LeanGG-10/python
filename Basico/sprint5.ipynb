Detección y tratamiento de outliers (es un valor extremadamente alto o bajo en comparación con la mayoría de los datos, y puede deberse a errores de medición, registro, o ser un valor real pero poco común.)

Se puede utilizar el rango interquartil (IQR) o simplemente un criterio.

```python
datos_practica.describe(include='all')
```

```python
#Calcular el rango interquartil de ingresos

Q1 = datos_practica['Ingreso'].quantile(0.25) #Obtenemos el primer quartil
Q3 = datos_practica['Ingreso'].quantile(0.75) #Obtenemos el tercer quartil

print(f'primer cuartil: {Q1}')
print(f'segundo cuartil: {Q3}')

#el rango solo es la reste entre el tercer cuartil y el primero

IQR = Q3 - Q1
print(f'rango interquartil: {IQR}')
```

```python
#Se define un rango aceptable para el análisis de datos

min_value = Q1 - 1.5 * IQR
max_value = Q3 + 1.5 * IQR

print(f'Los rangos aceptables son: valor mínimo ({min_value}) y valor máximo ({max_value})')
```

```python
#Filtramos valores fuera del rango para detectar los outliers 

outliers = (datos_practica['Ingreso'] < min_value) | (datos_practica['Ingreso'] > max_value) #Aquí definimos una máscara booleana de pandas (que a su ves es como un array o serie que almacena datos booleanos), cada fila del dataframe tiene un filtro de True o False (pero no afectan los datos) que está definido por las condiciones que le asignemos 

print(f'Los outliers serían: \n {datos_practica[outliers]}') #Aquí solo imprimimos las filas que se hayan filtrado como True
print(f'Valores booleanos dentro de "outliers":\n{outliers}') #Aquí se puede ver la serie o array con los tipos de datos correspondiente por cada fila existente en el dataframe, cada True representa un outlier ya que cumplió con la condición que se específico para ser considerado un outlier al crear la máscara, y cada False no representa un outlier ya que no cumplió con la condición para ser considerado un outlier
```

```python
#Corrección de outliers 

#Eliminarlos
datos_practica = datos_practica[~outliers] #Esto los elimina, el sigo '~' invierte todos los valores booleanos dentro de la máscara de pandas de outliers (Los True los vuelve False y viceversa), Se reasigna en el dataframe todas las filas filtrados por True dejando a los outliers fuera eliminándolos 

datos_practica
```

```python
#Corrección de outliers

#Reasignarle el valor mínimo o máximo según corresponda o algún otro valor que creamos mejor que sea reemplazarlo

datos_practica.loc[datos_practica['Ingreso'] < min_value, 'Ingreso'] = min_value #usamos .loc[] para seleccinar filas, columnas o casillas específicas dentro de los dataframes y a su vez poder reasginarles un nuevo valor 

datos_practica.loc[datos_practica['Ingreso'] > max_value, 'Ingreso'] = max_value

datos_practica

```
