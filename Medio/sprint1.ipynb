## Generación de gráficos básicos

```python
import matplotlib.pyplot as plt
import pandas as pd

df_practica = pd.read_excel('../datos/ventas_ejemplo.xlsx')
print(df_practica.dtypes)
df_practica
```

```python
#Gráfico de líneas: estos gráficos suelen usarle para ver los cambios a lo largo del tiempo


#Separar el mes y el año del dataframe

df_practica['Mes'] = pd.to_datetime(df_practica['Fecha']).dt.month #Obtenemos el mes de la columna fecha y la almacena en la nueva columna Mes, solo devuelve el número del mes
df_practica['Año'] = pd.to_datetime(df_practica['Fecha']).dt.year #Obtenemos el año de la columna fecha y la almacena en la nueva columna Año
df_practica
```

```python
#Eliminar signos de dólar y comas en una columna de strings

df_practica['Monto'] = df_practica['Monto'].str.replace('$', '') #Usamos la función .str.replace() para eliminar contenido de una columna que sea de tipo de dato string (object) al reemplazar con un string vacío
df_practica['Monto'] = df_practica['Monto'].str.replace(',', '') #.str.replace('Texto o contenido que queremos reemplazar', 'Texto o contenido por el que vamos a reemplazar lo anterior')
print(df_practica.dtypes) #Eliminamos cualquier tipo de signo dentro de la columna Monto para poder cambiar el tipo de la columna a float sin ningún problema o error
df_practica['Monto'] = df_practica['Monto'].astype(float)
df_practica.dtypes
```

```python
df_practica #Monto ya como float
```

```python
#Agrupar por mes y monto

#La función groupby() separa por grupos los datos que se repiten dentro de la columna que le damos como parámetro por eso es que el mes 4 que se repite 2 veces los separó en un grupo aparte de los demás y con la función .sum() suma todos los valores de este grupo y de todos los demás.

df_mesYmonto = df_practica.groupby('Mes')['Monto'].sum() #Creamos un nuevo dataframe que almacene los datos de esta agrupación, se usa la función .groupby('ponemos la columna del dataframe que queremos separar por grupos')['Ponemos la otra columna con la que queremos agrupar'].sum(), la función .sum() es importante porque es la acción que queremos hacer con los datos agrupados, si no le ponemos una acción no va a devolver nada
print(df_mesYmonto)
```

```python
lista_montos = df_mesYmonto.values  #Ponemos los valores de los meses en una variable para poder usar los datos en un gráfico (esto convierte a la variable que creamos en un array de numpy)

meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'] #usamos un diccionario para tener los meses de referencia

#mapear los índices de df_mesYmonto
meses_index = df_mesYmonto.index #Obtenemos los índices de los meses 
meses = [meses[i-1] for i  in meses_index] #selecciona los meses del diccionario según los índices que hayan en meses_index

print(meses)
plt.plot(meses, lista_montos, marker='o')
plt.title('Ventas por meses')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.show()



```

```python
#Agrupar monto por países y hacer un gráfico

df_ciudadYmonto = df_practica.groupby('Ciudad')['Monto'].sum()

plt.plot(df_ciudadYmonto.index, df_ciudadYmonto.values, marker='o')
plt.title('Ventas por ciudades')
plt.xlabel('Ciudades')
plt.ylabel('Ventas')
plt.show()
```

### Pasos para hacer un gráfico con dos columnas de datos 

- Crear un nuevo datframe y en este nuevo dataframe agrupar mediante la función groupby() las columnas que queramos usar en el gráfico.

- Obtener los valores de la agrupación.

- Obtener los índices de la agrupación.

- Usar los índices como columna, y los valores como filas (o como mejor nos convenga o lo permita el gráfico que se va a utilizar) en la función plt.plot() o en el gráfico que vayamos a utilizar.

- Usar la personalización de plt (como plt.title(), etc) que necesitemos en el gráfico.

- Mostrar la gráfica con plt.show() al final 

```python
#Gráfico de barras

#Para los gráficos de barra se usa la función plt.bar('columnas', 'filas') obligatorio poner los valores siempre como filas porque si no muestra una gráfica vacía
plt.bar(df_ciudadYmonto.index, df_ciudadYmonto.values) 
plt.title('Ventas por ciudades')
plt.xlabel('Ciudades')
plt.ylabel('Ventas')
plt.show()
```

```python
#En caso de que tengamos valores muy grandes éstos se van a mostrar probablemente como notación científica para evitar eso ponemos la siguiente línea de código

plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)  
```

## Gráficos con matplotlib

- Gráfico de líneas: plt.plot(columnas, filas, marker='o'), los valores pueden ir tanto como columnas o como filas y el atributo marker es opcional y tiene que ser uno compatible.

- Gráfico de barras: plt.bar(columnas, filas), los valores es obligatorio que vayan en las filas, caso contrario se muestra una gráfica vacía.

- Gráfico de pastel: plt.pie(valores, label='etiquetas o nombre representando a los valores', autopct='formato de porcentaje que representan los valores') los colores por ahora son aleatorios aunque se los puede configurar con el atributo colors

- Gráfico de frecuencias: plt.hist(valores que queramos ver su frecuencia por intervalos), este tipo de diagrama separa los datos en intervalos y muestra la frecuencia de datos dentro de esos intervalos.

- Diagramas de caja: plt.boxplot(valores que queramos ver su diagrama de caja)

- Diagrama de dispersión: plt.scatter(columnas, filas), los valores pueden ir tanto en las columnas como en las filas y separa los puntos en intervalos como los diagramas de frecuencias.



Los gráficos de frecuencias y diagramas de caja pueden comparar dos valores poniendo los valores entre corchetes plt.boxplot([valores 1, valores 2]) 

```python
#Gráfico con ciudad y cajas enviadas

df_ciudadYcajasenviadas =  df_practica.groupby('Ciudad')['Cajas Enviadas'].sum()

plt.bar(df_ciudadYcajasenviadas.index, df_ciudadYcajasenviadas.values)
plt.title('Ciudad y cajas enviadas')
plt.xlabel('Ciudades')
plt.ylabel('Cajas enviadas')
plt.show()
```

```python
#Gráfico de pastel con ciudad y cajas enviadas

plt.pie(df_ciudadYcajasenviadas.values, labels=df_ciudadYcajasenviadas.index, autopct='%1.1f%%')
plt.title('Ciudad y cajas enviadas')
plt.show()
print(df_ciudadYcajasenviadas)
```

```python
#Histograma de cajas enviadas

plt.hist(df_practica['Cajas Enviadas'])
plt.title('Cajas enviadas')
plt.show()
```

```python
#Diagramas de cajas de monto

plt.boxplot(df_practica['Monto'])
plt.title('Diagrama de cajas de Monto')
plt.show()
```

```python
#Combinar varios valores en un diagrama

plt.figure(figsize=(10, 6)) #con plt.figure(figsize=(ancho del diagrama, alto del diagrama)) podemos configurar el tamaño el diagrama que queramos ver
plt.boxplot([df_practica['Monto'], df_practica['Cajas Enviadas']]) #Solo se puede hacer con diagramas de caja y de dispersión
plt.title('Comparación de monto y cajas enviadas')
plt.show()
```

```python
plt.scatter(df_practica['Cajas Enviadas'], df_practica['Ciudad'])
plt.title('dispersión de cajas enviadas ')
plt.show()
```

## Gráficos con Seaborn

- Gráfico de dispersión: sb.scatterplot(x='nombre de columna del dataset con valores para las columnas', y='nombre de columna del dataset con valores para las filas', hue='columna para agrupar por categorías los colores', size=(ancho, alto), alpha='modifica transparencia de los datos', data=dataset o dataframe utilizado para el gráfico) debe haber por lo menos una columna con valores ya sea en las columna o fila del gráfico de dispersión

- Gráfico de dispersión con una línea de regresión: sb.lmplot(x='nombre de columna del dataset con valores para las columnas', y='nombre de columna del dataset con valores para las filas', hue='columna para agrupar por categorías los colores', fit_reg=(True o False) indica si queremos que sea visible la línea de regresión, alpha='modifica transparencia de los datos', lowess=(True o False) indica si queremos limitar la regresión basado en los valores de cada agrupación, row='nombre de columna por el que queramos ver gráficos separadas diferentes por categoría', data=dataset o dataframe utilizado para el gráfico)

- Gráficos histogramas: sb.histplot(columna el dataframe que queramos ver la gráfica del histograma, kde=(True o False) muestra una línea que se va curvando según el comportamiento de los valores, ) muestra los datos igual que el de matplotlib pero este permite más personalización.

- Gráfico de barras: sb.barplot(x='columna para la columna del gráfico', y='columna para las filas del gráfico', data=dataframe o dataset con los datos).

- Gráfico de caja: sb.boxplot(x='columna', y='valores', hue='columna para agrupar categorías con colores', data=fuente de datos) permite mostrar varios diagramas de caja según la cantidad de valores categóricos que tengamos en la columna que pasemos como parámetros en la x=''. 

- Gráfico de dispersión por categorías: sb.stripplot(x='columna', y='filas', data=fuente de datos, hue='columna para agrupar categorías con colores', jitter=(True(separa un poco los valores) o False(no separa los valores)), dodge=(True(separa los valores por las categorías en caso de que esten muy juntas) o False(no separa los valores por las categorías))).

- Gráfico todos con todos: sb.pairplot(fuente de datos, hue='columna para agrupar categorías con colores') empareja en pares todas las columnas con valores numéricos entre sí en gráficos de dispersión y hace una diagonal con gráfico de barras,al usar hue cambia el gráfico de barras por un gráfico de densidad. Este gráfico se demora en mostrarse.

```python
import matplotlib.pyplot as plt
import seaborn as sb

#Seaborn ya viene con datasets descargados para poder ser utilizados y prácticar 
sb.get_dataset_names() #Así obtenemos el nombre de los datasets de seaborn
```

```python
iris = sb.load_dataset('iris') #Se usa la función sb.load_dataset('nombre del dataset de seaborn') para cargar el dataset
print(iris.dtypes)
iris
```

```python
titanic = sb.load_dataset('titanic')

titanic
```

```python
sb.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.show()
```

```python
sb.lmplot(x='sepal_width', y='sepal_length', hue='species', row='species', data=iris)
plt.show()
```

```python
#Histogramas

sb.histplot(iris['petal_length'], kde=True)
plt.show()
```

```python
#Gráfico de barras

sb.barplot(data=titanic, x='sex', y='survived')
plt.show()
```

```python
#Diagramas de caja 

sb.boxplot(x='species', y='sepal_width', hue='species', data=iris)
plt.show()
```

```python
sb.boxplot(x='deck', y='age', hue='sex', data=titanic)
plt.show()
```

```python
sb.stripplot(x='species', y='sepal_length', hue='species', data=iris, jitter=True, dodge=True)
plt.show()
```

```python
sb.pairplot(iris)
plt.show()
```

```python
sb.pairplot(titanic, hue='sex')
plt.show()
```

```python
import plotly.express as px
import seaborn as sb 
import pandas as pd

#Mapa geográfico ejemplo titanic, para hacer este gráfico es importante tener siempre las coordenadas y conteo de datos para tener puntos de densidad



port_coordinates = {
    "Cherbourg":{"lat":49.633, "lon":1.6167},
    "Queenstown":{"lat":51.8494, "lon":-8.2946},
    "Southampton":{"lat":50.9097,"lon":-1.4044}
} 

pasajerosPorpuerto = titanic["embark_town"].value_counts().reset_index()
pasajerosPorpuerto.colums = ["embark_town", "count"]

pasajerosPorpuerto["lat"] = pasajerosPorpuerto["embark_town"].map(lambda x:port_coordinates[x]["lat"])
pasajerosPorpuerto["lon"] = pasajerosPorpuerto["embark_town"].map(lambda x:port_coordinates[x]["lon"])

#Crear mapa geográfico

fig = px.scatter_geo(pasajerosPorpuerto,
                    lat="lat", lon="lon",
                    size="count", hover_name="embark_town",
                    title="Pasajeros por puerto",
                    projection="natural earth")
fig.show()

```
