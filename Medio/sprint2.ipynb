# Correlación

```python

import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


iris = sb.load_dataset('iris')
titanic = sb.load_dataset('titanic')
```

```python
#Matriz de correlación

sb.heatmap(iris.drop(columns=['species']).corr(), annot=True) #se borra la columna categórica porque heatmap no permite datos que no sean numéricos, el annot=True es para que el gráfico salga con el número de la correlación

plt.show()
```

```python
#Seleccionar solo variables númericas de un dataset

titanicNumeros = titanic.select_dtypes(include=['int64', 'float64']) #con la función selec_dtypes podemos seleccionar los tipo de datos que queramos dentro del dataframe, los tipos que datos que queremos los ponemos dentro de include=['tipo de dato a seleccionar']

sb.heatmap(titanicNumeros.corr(), annot=True)
plt.show()
```

# Funciones de personalización de gráficos de Matplotlib

- plt.figure(figsize=(ancho,alto)) = cambia el tamaño de la figura
- fig, ax = plt.subplots(figsize(ancho, alto)) = Crea una figura y ejes con tamaño específico
- ax.plot(x, y, color='red', linestyle='--', linewidth=2, marker='o') = Personaliza color, estilo de línea, grosor y marcadores
- ax.set_title('Titulo', fontsize=14, fontweight='bold', color='blue') = Personaliza el título del gráfico
- ax.set_xlabel('Etiqueta x', fontsize=12) = Establece la etiqueta del eje x
- ax.set_ylabel('Etiqueta y', fontsize=12) = Establece la etiqueta del eje y
- ax.set_xticks([0, 2, 4, 6, 8, 10]) = Personaliza los ticks del eje x
- ax.set_yticks([-1, 0, 1]) = Personaliza los ticks del eje y
- ax.legend(loc='upper right') = Muestra la leyenda en la ubicación que se indica
- ax.grid(True, linestyle=':', linewidth=0.7) = Agrega una cuadrícula con estilo personalizado
- ax.set_facecolor('Color en código') = Cambia el color del fondo del gráfico
- plt.xticks(rotation=45) = Rota las etiquetas del eje x
- plt.yticks(rotation=90) = Rota las etiquetas del eje y


```python
#Creamos un gráfico con tamaño específico
plt.figure(figsize=(5,5))

#Creamos un gráfico de barras
plt.bar(df['species'].value_counts().index, df['species'].value_counts().values, color='red')

#Personalizamos el gráfico
plt.title('DISTRIBUCION DE ESPECIES EN EL DATASET IRIS', fontsize=16, fontweight='bold')
plt.xlabel('especies', fontsize=12)
plt.ylabel('frecuencia', fontsize=14)
plt.gca().set_facecolor('#ffff00')
plt.gca().set



plt.show()
```

```python
#Crear un subplot: subplot es una función que permite crear varios gráficos dentro de una misma figura
#fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,5)) 


#Si se aumenta el número de filas o columnas, se deben agregar más variables para cada gráfico, por ejemplo: 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15,10))

#Primer subplot
ax1.scatter(df['sepal_length'], df['sepal_width'], color='white')
ax1.set_title('SEPAL LENGHT VS SEPAL WIDTH', fontsize=16, fontweight='bold')
ax1.set_xlabel('sepal length', fontsize=12)
ax1.set_ylabel('sepal width', fontsize=10)
ax1.grid(True, linestyle=':', linewidth=0.7)
ax1.set_facecolor('#0022ff')

#Segundo subplot
ax2.scatter(df['petal_length'], df['petal_width'], color='red')
ax2.set_title('PETAL LENGHT VS PETAL WIDTH', fontsize=16, fontweight='bold')
ax2.set_xlabel('petal length', fontsize=12)
ax2.set_ylabel('petal width', fontsize=10)
ax2.grid(True, linestyle='--', linewidth=0.7)
ax2.set_facecolor("#17B611")

#Tercer subplot 
ax3.hist(df['sepal_length'], bins=20, color='black')
ax3.set_title('SEPAL LENGHT', fontsize=20, fontweight='bold')
ax3.set_xlabel('sepal length', fontsize=12)
ax3.set_ylabel('frecuencia', fontsize=10, color='green')
ax3.set_facecolor('#ff00ff')

#Cuarto subplot
ax4.hist(df['petal_length'], bins=20, color='cyan')

plt.tight_layout() #Con esta función ajustamos el espacio entre los subplots pa que no se solapen entre sí
plt.show()

```

```python
#Configurar los valores de los ejes
plt.figure(figsize=(10,3))

plt.scatter(df['sepal_length'], df['sepal_width'], color='green')
plt.title('SEPAL LENGHT VS SEPAL WIDTH', fontsize=16, fontweight='bold')
plt.xlabel('sepal length', fontsize=12)
plt.ylabel('sepal width', fontsize=10)
plt.grid(True, linestyle=':', linewidth=0.7)
plt.gca().set_facecolor('#0022ff')

#Modificamos los valores de los ejes
plt.xlim(4, 10) #Modificamos el rango de valores que mostraremos en un eje, en este caso el eje x
plt.ylim(2, 6) #Modificamos el rango de valores que mostraremos en un eje, en este caso el eje y
#Establecemos los valores que queremos mostrar en los ejes
plt.gca().set_xticks([4, 5, 7, 8, 10])
plt.gca().set_yticks([2, 3, 6])
#personalizamos los ticks
plt.xticks(fontsize=20, color='orange')
plt.yticks(fontsize=15, color='purple')

plt.show()
```

# Funciones de personalización de gráficos de seaborn

- sb.set_style('whitegrid') = cambia el estilo del gráfico
- sb.set_palette('pastel') = cambia la paleta de colores del gráfico
- ax = sb.barplot(x='col_x', y='col_y', data=df, edgecolor='black') = crear gráfico de barras con borde
- ax.set_title('titulo', fontsize=10, fontweight='bold', color='red') = personaliza titulo del gráfico
- ax.set_xlabel('etiqueta x', fontsize=12) = personaliza etiqueta del eje x
- ax.set_ylabel('etiqueta y', fontsize=12) = personaliza etiqueta del eje y
- ax.set_facecolor('codigo del color') = cambia el color de fondo del gráfico
- plt.xticks(fontsize=10, color='blue') = cambia el tamaño y color de los ticks del eje x
- plt.yticks(fontsize=10, color='blue') = cambia el tamaño y color de los ticks del eje y

```python
sb.set_style('whitegrid')
sb.set_palette('pastel')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 5))

#Primer subplot
sb.scatterplot(x='sepal_length', y='sepal_width', color='white', data=df, ax=ax1)
ax1.set_title('SEPAL LENGHT VS SEPAL WIDTH', fontsize=16, fontweight='bold')
ax1.set_xlabel('sepal length', fontsize=12)
ax1.set_ylabel('sepal width', fontsize=10)
ax1.set_xticks([4, 5, 6, 7, 8, 9 ,10])
ax1.set_yticks([2, 3, 4, 5, 6])
ax1.grid(True, linestyle=':', linewidth=0.7)
ax1.set_facecolor('#0022ff')

#Segundo subplot
sb.scatterplot(x='petal_length', y='petal_width', color='black', data=df, ax=ax2)
ax2.set_title('PETAL LENGHT VS PETAL WIDTH', fontsize=16, fontweight='bold')
ax2.set_xlabel('petal length', fontsize=12)
ax2.set_ylabel('petal width', fontsize=10)
ax2.set_xticks([1, 2, 3, 4, 5, 6])
ax2.set_yticks([0.5, 1, 1.5, 2, 2.5, 3])
ax2.grid(True, linestyle=':', linewidth=0.7)
ax2.set_facecolor("#c1df1b")

#Los auxiliares ax tienen los mismos métodos de personalización que matplotlib, por lo que se puede personalizar cada subplot de la misma manera que personalizamos un gráfico normal.
plt.tight_layout()
plt.show()
```
