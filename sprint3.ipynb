# Limpieza de datos

```python
#Consiste en identificar y corregir o eliminar datos que sean incorrectos, incompletos, duplicados, irrelevantes o que estén fuera de lugar dentro de un conjunto de datos
#Antes de hacer cualquier proceso siempre se debe hacer una limpieza de datos
#Esta es la parte más importante al trabajar con datos 

import pandas as pd

mff = pd.read_excel('datos/mff.xlsx')

mff.describe()
```

- count: números de valores encontrados en la columna
- mean: media de los valores encontrados en la columna
- std: desviación estándar de los valores encontrados en la columna
- min: valor mínimo de los valores encontrados en la columna
- 25%: primer quartil de los valores encontrados en la columna
- 50%: mediana de los valores encontrados en la columna
- 75%: tercer quartil de los valores encontrados en la columna
- max: valor máximo de los valores encontrados en la columna

```python
#Manejo de datos duplicados

#Conseguir datos duplicados

print(mff['id'].duplicated().sum()) #imprime la cantidad de datos duplicados que hay en la columna seleccionada (el índice/nombre de la columna, es obligatoria)

mff[mff["id"].duplicated()] #imprime los datos que se duplican pero solo muestra el valor duplicado que esté más completo, el resto de valores duplicados no los muestra
```

```python
mff[mff["id"] == 24] #imprime todos los valores que cumpla una condición (si hay valores duplicados podemos ver todos los duplicados así)
```

```python
#borrar datos duplicados

#Los datos duplicados solo se borran en el dataframe, el archivo o origen de los datos queda intacto
mff.drop_duplicates(subset='id', keep = 'last', inplace = True) #para borrar duplicados se llama a la función drop_duplicates(subset = 'índice o columna en que vamos a aplicar la función', keep ='elegimos el valor duplicado que queramos mantener (en este caso last sería el último, aplica para todos los duplicados)', inplace = True (si ponemos true modifica el dataframe original, si ponemos false no modifica el dataframe original sino que crea una copia con los cambios realizados (en este caso eliminar datos duplicados) dejando el dataframe original intacto)

mff 
```

```python
print(mff.isnull().sum()) #imprime el nombre de las columnas más la cantidad de valores nulos que poseen
print("=========")
print(mff.dtypes) #imprime el nombre de las columnas más el tipo de datos de éstas 
```

```python
mff
```

```python
#Rellenar campos vacíos 

#Rellenar usando la media 
mff["Precio"].fillna(mff["Precio"].mean(), inplace=True) #Usamos la función fillna(columna que queramos rellenar.Dato que queramos usar como relleno 'toma el dato de la columna', si queremos modificar el dataframe o hacer los datos creando otro dataframe)

#Rellenar usando la moda 
mff["Tier"].fillna(mff["Tier"].mode()[0], inplace=True) #la función mode()[si solo hay una moda se usa el índice 0, pero si hay más de una moda esa moda se va acumulando entro de la función y podemos cambiar el índice según cual esté ocupanod la moda que queramos utilizar] sirve para sacar la moda de un conjunto de datos, si no le ponemos un índice a la función no va a rellenar nada xq la función fillna solo permite valores únicos

mff["id"].fillna(mff["id"].mean(), inplace = True)

mff["Nivel"].fillna(mff["Nivel"].mean(), inplace = True)
```

```python
#Rellenar campos con string
mff["Personajes"].fillna('No hay personaje', inplace=True)
mff
```

```python
print(mff.isnull().sum())
print("========================")
#cambiar tipo de dato de un columna 
mff["Tier"] = mff["Tier"].astype(int) #para cambiar el tipo de dato dentro de un dataframe escogemos la columna que queramos cambiar el dato y la reasignamos un nuevo tipo de dato con la función astype('tipo de dato que queramos poner')

mff["Nivel"] = mff["Nivel"].astype(int)

print(mff.dtypes)
```

```python
#Normalización
#Para aplicar la normalización primer se debe instalar la librería "pip install scikit-learn"

import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler 

print("Dataframe original ==================================")
print(mff)
print("Dataframe sin tipo de dato string ======================================")

mff_sinString = mff.drop(columns=["Personajes"]) #Borramos la columna con string de nuestro dataframe porque la normalización no funciona con otros tipos de datos que no sean numéricos
print(mff_sinString)
print("Normalizacion minMax ======================================")
#Normalización MinMax
nomalizacion_minMax = MinMaxScaler((0,10))  #Guardamos la función con los valores mínimo y máximos que queremos normalizar en una nueva variable para facilitar su uso (en sintaxis)
mff_minMax = nomalizacion_minMax.fit_transform(mff_sinString) #Guardamos en una nueva variable los datos normalizados más que todo para después poder comparar y entender cómo funciona. También llamamos la función fit_transform(datos que queremos normalizar)
print(mff_minMax)
print("Normalización Z-Score ========================================")

#Normalización Z-score
normalizacion_zScore = StandardScaler() #Guardamos la función en una nueva variable 
mff_zScore = normalizacion_zScore.fit_transform(mff_sinString) 
print(mff_zScore)
print("Normalizacion una columna =================================================")
#Si tenemos un dataframe con strings y solo queremos normalizar una columna que no sea string no es necesario crear un nuevo dataframe sin string como hice primero. También se puede normalizar columna por columna y no todas las columnas a la vez

minMaxOneColumn = nomalizacion_minMax.fit_transform(mff[["Precio"]]) #debemos poner el dataframe y para elegir la columna debemos poner la columna que queremo normalizar dentro de dos corchetes seguidos [["Columna a normalizar dentro el dataframe"]]
print(minMaxOneColumn)




```

```python
#La función de fit_transform() devuelve un array, no un dataframe, por lo que si queremos usar los nuevos datos normalizados con las funciones de pandas hay que tranformarlos en dataframe

#Transformar array en dataframe

mffzScore_dataframe = pd.DataFrame(mff_zScore, columns=mff_sinString.columns)
#print(mffzScore_dataframe)
print(mffzScore_dataframe.describe())
print("=======================================")
minmaxMff_dataframe = pd.DataFrame(mff_minMax, columns = mff_sinString.columns)
print(minmaxMff_dataframe.describe())
```
